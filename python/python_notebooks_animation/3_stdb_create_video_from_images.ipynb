```python
#Function, create a video from the images

import os
import cv2
import re

def create_video(input_folder, output_path, fps):
    images = []

    # Regular expression to extract the number after the underscore (_)
    def extract_number_after_underscore(filename):
        match = re.search(r'_(\d+)', filename)  # Look for a number after the underscore
        return int(match.group(1)) if match else float('inf')  # Default to infinity if no match found

    # Sort files based on the extracted number
    for filename in sorted(os.listdir(input_folder), key=extract_number_after_underscore):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(input_folder, filename)
            images.append(image_path)
    
    if not images:
        print("No images found in the input folder.")
        return
    
    # Read the first image to get the dimensions
    frame = cv2.imread(images[0])
    height, width, layers = frame.shape

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # You can use other codecs like 'XVID'
    video = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    for image in images:
        frame = cv2.imread(image)
        video.write(frame)

    video.release()
    print(f"Video saved at {output_path}")
```

```python
input_folder = 'D:\\routes_animation2_12'
fps = 12
output_path = 'C:\\Users\\andre\\OneDrive\\Dokumente\\UNI\\Master_Digital_Humanities\\S3_WS2024\\LP_Data_Analysis_Project\\maps\\routes_animation2_date.mp4'
create_video(input_folder, output_path, fps)
```
